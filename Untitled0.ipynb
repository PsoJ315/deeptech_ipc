```python
# =========================================
# 0. 코랩 환경 설정
# =========================================
!apt-get -qq update
!apt-get -qq install -y fonts-nanum
!pip -q install kiwipiepy scikit-learn openpyxl

import os
import re
import glob
import math
from collections import Counter, defaultdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from kiwipiepy import Kiwi
from sklearn.feature_extraction.text import TfidfVectorizer

plt.rcParams["font.family"] = "NanumGothic"
plt.rcParams["axes.unicode_minus"] = False

# =========================================
# 1. 설정
# =========================================
INPUT_GLOB = "/content/drive/MyDrive/deeptech_ipc/세부분석/20260314132243.csv"
OUTPUT_DIR = "/content/output_term_extraction"
os.makedirs(OUTPUT_DIR, exist_ok=True)

DEEPTECH_CODES = {
    "A61",
    "B01","B03","B04","B05","B06","B07","B08","B09","B21","B22","B23","B24","B25","B26","B27","B28","B29","B30","B31","B32","B41","B44","B60","B61","B62","B63","B64","B65","B66","B67","B81","B82",
    "C01","C02","C03","C04","C05","C06","C07","C08","C09","C10","C11","C12","C13","C21","C22","C23","C25","C30","C40",
    "D01","D06","D21",
    "E01","E02","E03","E04","E05","E06","E21",
    "F01","F02","F03","F04","F15","F16","F17","F21","F22","F23","F24","F25","F26","F27","F28",
    "G01","G02","G03","G05","G06","G07","G08","G09","G11","G16","G21",
    "H01","H02","H03","H04","H05","H10"
}

# 특허 상투어 / 분석 노이즈
GENERIC_STOPWORDS = {
    "장치","방법","시스템","구성","포함","포함하는","포함하며","사용","이용","제공","처리","제어",
    "발명","특허","상기","하나","이상","이하","복수","복수의","적어도","관련","형성","형성된",
    "기술","데이터","정보","대한","관한","이용한","통한","통해","경우","의해","의한","위한",
    "에서","으로","하는","되는","있다","한다","가능","결과","구비","구비한","구비하는",
    "각각","부분","모듈","단계","영역","값","부","유닛","기반","서비스","명칭","요약","청구항",
    "출원","등록","공개","문헌","실시예","적용","특징","효과","목적","형태","구조","동작",
    "대상","범위","부재","단말","기기","장치부","처리부","제어부","사용자","서버","클라이언트"
}

# 너무 일반적인 한 글자/기호 제외
MIN_TERM_LEN = 2
MIN_DF = 2           # 최소 문헌 수
MIN_TF = 3           # 최소 총출현 빈도
MAX_NGRAM = 4        # 최대 복합어 길이

# 기술어 후보가 될 만한 태그
# NNG: 일반명사, NNP: 고유명사, SL: 영문, SN: 숫자, XR: 어근
VALID_TAG_PREFIX = ("NNG", "NNP", "SL", "SN", "XR")

# 복합어 내부에서 허용할 연결 태그
# XPN: 체언접두사, XSN: 명사파생접미사
CONNECTOR_TAG_PREFIX = ("XPN", "XSN")

kiwi = Kiwi()

# =========================================
# 2. CSV 로딩 및 전처리
# =========================================
def normalize_text(x):
    if pd.isna(x):
        return ""
    x = str(x)
    x = x.replace("\n", " ").replace("\r", " ").replace("\t", " ")
    x = re.sub(r"\s+", " ", x)
    return x.strip()

def parse_year(date_str):
    if pd.isna(date_str):
        return np.nan
    s = str(date_str).strip()
    m = re.match(r"^(\d{4})", s)
    return int(m.group(1)) if m else np.nan

def extract_ipc_codes(ipc_raw):
    if pd.isna(ipc_raw):
        return []
    codes = re.findall(r"([A-H]\d{2})", str(ipc_raw))
    seen = set()
    out = []
    for c in codes:
        if c not in seen:
            out.append(c)
            seen.add(c)
    return out

def read_kipris_csv(path):
    try:
        df = pd.read_csv(path, skiprows=7, encoding="utf-8-sig", engine="python")
    except UnicodeDecodeError:
        df = pd.read_csv(path, skiprows=7, encoding="cp949", engine="python")
    df["source_file"] = os.path.basename(path)
    return df

def pick_core_columns(df):
    wanted = [
        "발명의명칭", "IPC분류", "출원번호", "출원일자", "출원인",
        "요약", "청구항", "등록번호", "등록일자", "공개번호", "공개일자"
    ]
    existing = [c for c in wanted if c in df.columns]
    return df[existing + ["source_file"]].copy()

def get_period(year):
    if pd.isna(year):
        return "unknown"
    year = int(year)
    if 2015 <= year <= 2019:
        return "2015-2019"
    elif 2020 <= year <= 2021:
        return "2020-2021"
    elif 2022 <= year <= 2024:
        return "2022-2024"
    return "other"

files = sorted(glob.glob(INPUT_GLOB))
files = [f for f in files if os.path.basename(f).endswith(".csv") and "output_" not in f]

if not files:
    raise FileNotFoundError("입력 CSV를 찾지 못했습니다. /content 에 KIPRIS CSV를 업로드하세요.")

print("읽을 파일 수:", len(files))
for f in files:
    print("-", os.path.basename(f))

raw_list = [read_kipris_csv(f) for f in files]
raw_df = pd.concat(raw_list, ignore_index=True)
df = pick_core_columns(raw_df)

for col in ["발명의명칭", "IPC분류", "출원인", "요약", "청구항"]:
    if col in df.columns:
        df[col] = df[col].apply(normalize_text)

df["출원연도"] = df["출원일자"].apply(parse_year)
df["IPC코드목록"] = df["IPC분류"].apply(extract_ipc_codes)
df["IPC대표"] = df["IPC코드목록"].apply(lambda x: x[0] if x else np.nan)
df["섹션"] = df["IPC대표"].apply(lambda x: x[0] if isinstance(x, str) else np.nan)
df["period"] = df["출원연도"].apply(get_period)

df["is_deeptech"] = df["IPC코드목록"].apply(lambda codes: any(c in DEEPTECH_CODES for c in codes))
df = df[df["is_deeptech"]].copy()

df["text_all"] = (
    df.get("발명의명칭", "").fillna("").astype(str) + " " +
    df.get("요약", "").fillna("").astype(str) + " " +
    df.get("청구항", "").fillna("").astype(str)
).apply(normalize_text)

if "출원번호" in df.columns:
    before = len(df)
    df = df.drop_duplicates(subset=["출원번호"]).copy()
    print("중복 제거:", before - len(df))

print("전처리 후 shape:", df.shape)
print("출원연도 범위:", df["출원연도"].min(), "~", df["출원연도"].max())

# =========================================
# 3. 정교한 기술어 추출기
# =========================================
def is_valid_term_token(token):
    tag = token.tag
    form = token.form.strip().lower()

    if len(form) < 1:
        return False
    if any(tag.startswith(p) for p in VALID_TAG_PREFIX):
        return True
    if any(tag.startswith(p) for p in CONNECTOR_TAG_PREFIX):
        return True
    return False

def clean_term(term):
    term = term.strip().lower()
    term = re.sub(r"\s+", " ", term)
    term = re.sub(r"^[^\w가-힣]+|[^\w가-힣]+$", "", term)

    if len(term) < MIN_TERM_LEN:
        return ""
    if term in GENERIC_STOPWORDS:
        return ""
    if re.fullmatch(r"[0-9]+", term):
        return ""
    if re.fullmatch(r"[a-z]", term):
        return ""
    return term

def term_token_sequence(text):
    """
    유효한 기술어 후보 토큰 시퀀스 추출
    """
    toks = kiwi.tokenize(text)
    seq = []
    for t in toks:
        if is_valid_term_token(t):
            seq.append((t.form.lower(), t.tag))
        else:
            seq.append(("", "BREAK"))
    return seq

def extract_candidate_terms(text, max_ngram=MAX_NGRAM):
    """
    한 문서에서 기술어 후보 추출
    - 단일 기술어
    - 복합 명사구
    """
    seq = term_token_sequence(text)
    terms = []

    # BREAK 기준으로 연속 구간 분리
    chunks = []
    current = []
    for form, tag in seq:
        if tag == "BREAK":
            if current:
                chunks.append(current)
                current = []
        else:
            current.append((form, tag))
    if current:
        chunks.append(current)

    for chunk in chunks:
        words = [w for w, _ in chunk]
        tags = [t for _, t in chunk]

        # 단일어
        for w, tag in chunk:
            term = clean_term(w)
            if term and term not in GENERIC_STOPWORDS:
                terms.append(term)

        # 복합어 n-gram
        n = len(words)
        for i in range(n):
            for j in range(i + 2, min(n, i + max_ngram) + 1):
                cand_words = words[i:j]
                cand_tags = tags[i:j]

                # 끝은 실질 명사/영문/숫자여야 함
                if not any(cand_tags[-1].startswith(p) for p in VALID_TAG_PREFIX):
                    continue

                # 복합어 내부에 의미 없는 접사만으로 끝나면 제외
                if all(t.startswith("X") for t in cand_tags):
                    continue

                cand = " ".join(cand_words)
                cand = clean_term(cand)
                if not cand:
                    continue

                # 일반 불용어만으로 이뤄진 경우 제외
                if all(w in GENERIC_STOPWORDS for w in cand.split()):
                    continue

                terms.append(cand)

    return terms

def extract_doc_terms(text):
    """
    한 문서에서 중복 없는 기술어 세트
    DF 계산용
    """
    return set(extract_candidate_terms(text))

# =========================================
# 4. 문헌별 기술어 추출
# =========================================
doc_terms_list = []
doc_term_counter = []
for idx, row in df.iterrows():
    terms = extract_candidate_terms(row["text_all"])
    term_set = set(terms)
    doc_terms_list.append(term_set)
    doc_term_counter.append(Counter(terms))

df["doc_terms"] = doc_terms_list

# 전체 TF / DF 계산
tf_counter = Counter()
df_counter = Counter()

for c in doc_term_counter:
    tf_counter.update(c)

for s in doc_terms_list:
    df_counter.update(s)

N_DOCS = len(df)

term_stats_rows = []
for term, tf in tf_counter.items():
    dfreq = df_counter[term]
    if tf < MIN_TF or dfreq < MIN_DF:
        continue

    # 기술어 점수:
    # TF-IDF 스타일 + 복합어 보너스
    idf = math.log((N_DOCS + 1) / (dfreq + 1)) + 1
    token_len = len(term.split())
    char_len = len(term.replace(" ", ""))
    len_bonus = 1.0 + 0.15 * max(0, token_len - 1) + 0.02 * max(0, char_len - 2)
    score = tf * idf * len_bonus

    term_stats_rows.append({
        "term": term,
        "tf": tf,
        "df": dfreq,
        "idf": round(idf, 4),
        "token_len": token_len,
        "char_len": char_len,
        "score": round(score, 4)
    })

term_stats = pd.DataFrame(term_stats_rows).sort_values(["score", "tf"], ascending=False).reset_index(drop=True)

# =========================================
# 5. 그룹별 기술어 추출
# =========================================
def group_term_stats(df, group_col):
    rows = []
    for group_name, sub in df.groupby(group_col):
        g_tf = Counter()
        g_df = Counter()
        n_docs = len(sub)

        for terms_set, text in zip(sub["doc_terms"], sub["text_all"]):
            g_df.update(terms_set)
            g_tf.update(extract_candidate_terms(text))

        for term, tf in g_tf.items():
            dfreq = g_df[term]
            if tf < 2 or dfreq < 2:
                continue

            idf = math.log((n_docs + 1) / (dfreq + 1)) + 1
            token_len = len(term.split())
            char_len = len(term.replace(" ", ""))
            len_bonus = 1.0 + 0.15 * max(0, token_len - 1) + 0.02 * max(0, char_len - 2)
            score = tf * idf * len_bonus

            rows.append({
                group_col: group_name,
                "term": term,
                "tf": tf,
                "df": dfreq,
                "idf": round(idf, 4),
                "score": round(score, 4)
            })

    out = pd.DataFrame(rows)
    if len(out) == 0:
        return out
    out = out.sort_values([group_col, "score", "tf"], ascending=[True, False, False])
    return out.reset_index(drop=True)

term_by_section = group_term_stats(df, "섹션")
term_by_ipc = group_term_stats(df, "IPC대표")
term_by_period = group_term_stats(df, "period")
term_by_year = group_term_stats(df, "출원연도")

# =========================================
# 6. TF-IDF 특징 기술어
# =========================================
def tfidf_terms_by_group(df, group_col, top_n=20):
    temp = df[[group_col, "text_all"]].dropna().copy()
    docs = temp.groupby(group_col)["text_all"].apply(lambda x: " ".join(x.astype(str))).reset_index()

    def analyzer(text):
        return extract_candidate_terms(text)

    vectorizer = TfidfVectorizer(
        analyzer=analyzer,
        lowercase=False,
        min_df=1,
        max_df=0.85
    )

    X = vectorizer.fit_transform(docs["text_all"])
    feature_names = np.array(vectorizer.get_feature_names_out())

    rows = []
    for i, group_name in enumerate(docs[group_col]):
        row = X[i].toarray().flatten()
        top_idx = row.argsort()[::-1][:top_n]
        for idx in top_idx:
            if row[idx] <= 0:
                continue
            rows.append({
                group_col: group_name,
                "term": feature_names[idx],
                "tfidf": float(row[idx])
            })
    return pd.DataFrame(rows)

tfidf_by_section = tfidf_terms_by_group(df, "섹션", top_n=20)
tfidf_by_ipc = tfidf_terms_by_group(df, "IPC대표", top_n=20)
tfidf_by_period = tfidf_terms_by_group(df, "period", top_n=20)

# =========================================
# 7. 기초 집계
# =========================================
yearly_counts = (
    df.groupby("출원연도", as_index=False)
      .size()
      .rename(columns={"size": "출원건수"})
      .sort_values("출원연도")
)

ipc_counts = (
    df.groupby("IPC대표", as_index=False)
      .size()
      .rename(columns={"size": "건수"})
      .sort_values("건수", ascending=False)
)

exploded = df[["출원번호", "출원연도", "period", "IPC코드목록"]].explode("IPC코드목록").rename(columns={"IPC코드목록": "IPC코드"})
exploded = exploded.dropna(subset=["IPC코드"]).copy()
exploded["섹션"] = exploded["IPC코드"].str[0]
exploded = exploded[exploded["IPC코드"].isin(DEEPTECH_CODES)].copy()

section_counts = (
    exploded.groupby("섹션", as_index=False)
            .size()
            .rename(columns={"size": "건수"})
            .sort_values("건수", ascending=False)
)
section_counts["비중(%)"] = section_counts["건수"] / section_counts["건수"].sum() * 100

# =========================================
# 8. 저장
# =========================================
clean_cols = [
    "출원번호", "출원일자", "출원연도", "period", "출원인",
    "발명의명칭", "요약", "청구항",
    "IPC분류", "IPC대표", "섹션", "text_all", "source_file"
]
clean_cols = [c for c in clean_cols if c in df.columns]

df[clean_cols].to_csv(f"{OUTPUT_DIR}/clean_patents.csv", index=False, encoding="utf-8-sig")
exploded.to_csv(f"{OUTPUT_DIR}/patents_ipc_exploded.csv", index=False, encoding="utf-8-sig")

yearly_counts.to_csv(f"{OUTPUT_DIR}/yearly_counts.csv", index=False, encoding="utf-8-sig")
ipc_counts.to_csv(f"{OUTPUT_DIR}/ipc_counts.csv", index=False, encoding="utf-8-sig")
section_counts.to_csv(f"{OUTPUT_DIR}/section_counts.csv", index=False, encoding="utf-8-sig")

term_stats.to_csv(f"{OUTPUT_DIR}/term_stats_overall.csv", index=False, encoding="utf-8-sig")
term_by_section.to_csv(f"{OUTPUT_DIR}/term_stats_by_section.csv", index=False, encoding="utf-8-sig")
term_by_ipc.to_csv(f"{OUTPUT_DIR}/term_stats_by_ipc.csv", index=False, encoding="utf-8-sig")
term_by_period.to_csv(f"{OUTPUT_DIR}/term_stats_by_period.csv", index=False, encoding="utf-8-sig")
term_by_year.to_csv(f"{OUTPUT_DIR}/term_stats_by_year.csv", index=False, encoding="utf-8-sig")

tfidf_by_section.to_csv(f"{OUTPUT_DIR}/tfidf_terms_by_section.csv", index=False, encoding="utf-8-sig")
tfidf_by_ipc.to_csv(f"{OUTPUT_DIR}/tfidf_terms_by_ipc.csv", index=False, encoding="utf-8-sig")
tfidf_by_period.to_csv(f"{OUTPUT_DIR}/tfidf_terms_by_period.csv", index=False, encoding="utf-8-sig")

# Datawrapper용 wide
year_section = (
    exploded.groupby(["출원연도", "섹션"])
            .size()
            .reset_index(name="건수")
            .pivot(index="출원연도", columns="섹션", values="건수")
            .fillna(0)
            .astype(int)
            .reset_index()
)
year_section.to_csv(f"{OUTPUT_DIR}/dw_year_section.csv", index=False, encoding="utf-8-sig")

top10_ipc = ipc_counts.head(10)["IPC대표"].tolist()
year_top_ipc = (
    df[df["IPC대표"].isin(top10_ipc)]
      .groupby(["출원연도", "IPC대표"])
      .size()
      .reset_index(name="건수")
      .pivot(index="출원연도", columns="IPC대표", values="건수")
      .fillna(0)
      .astype(int)
      .reset_index()
)
year_top_ipc.to_csv(f"{OUTPUT_DIR}/dw_year_top10_ipc.csv", index=False, encoding="utf-8-sig")

print("\n저장 완료:", OUTPUT_DIR)

# =========================================
# 9. 요약 출력
# =========================================
print("\n[전체 기술어 상위 30]")
print(term_stats.head(30))

print("\n[섹션별 기술어 상위 예시]")
print(term_by_section.groupby("섹션").head(15).head(120))

print("\n[IPC대표별 TF-IDF 기술어 예시]")
print(tfidf_by_ipc.groupby("IPC대표").head(15).head(150))

# =========================================
# 10. 시각화
# =========================================
# 10-1 전체 상위 기술어
top_terms = term_stats.head(20).copy()
plt.figure(figsize=(10, 8))
bars = plt.barh(top_terms["term"][::-1], top_terms["score"][::-1])
plt.title("전체 상위 기술어 Top 20 (score 기준)")
plt.xlabel("score")
plt.ylabel("기술어")
for b, v in zip(bars, top_terms["score"][::-1]):
    plt.text(b.get_width() + max(top_terms["score"]) * 0.01, b.get_y() + b.get_height()/2, f"{v:.1f}", va="center")
plt.tight_layout()
plt.show()

# 10-2 연도별 출원 건수
plt.figure(figsize=(10, 5))
plt.plot(yearly_counts["출원연도"], yearly_counts["출원건수"], marker="o", linewidth=2)
plt.title("연도별 출원 건수")
plt.xlabel("출원연도")
plt.ylabel("건수")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 10-3 섹션 비중
plt.figure(figsize=(9, 5))
bars = plt.bar(section_counts["섹션"], section_counts["비중(%)"])
plt.title("섹션별 비중")
plt.xlabel("섹션")
plt.ylabel("비중 (%)")
for b, v in zip(bars, section_counts["비중(%)"]):
    plt.text(b.get_x() + b.get_width()/2, b.get_height() + 0.2, f"{v:.1f}%", ha="center", va="bottom", fontsize=9)
plt.tight_layout()
plt.show()
```
